```python
# read json file
with open('./DiscourseSimplification/output.json', 'r') as f:
    data = json.load(f)
```

```python
data['sentences'][0]['elementMap'].keys()
```

```python
data['sentences'][0]['elementMap']['527968a399b44295941ea31303ebc337']['text']
```

```python
dissim_dict = {}
for i in range(len(data['sentences'])) :
    dissim_dict[data['sentences'][i]['originalSentence']] = [data['sentences'][i]['elementMap'][key]['text'] for key in data['sentences'][i]['elementMap'].keys()]
```

```python
dissim_dict
```

```python
with open('../data/dissim_dict.json', 'w') as f:
    json.dump(dissim_dict, f)
```

```python
f = open('../data/10/output_', 'w')
for i in range(100) :
    f.write('# ' + data['sentences'][i]['originalSentence'] + '\n')
    for j, key in enumerate(data['sentences'][i]['elementMap'].keys()):
        f.write('\t' + str(j+1) + ' ' + data['sentences'][i]['elementMap'][key]['text'] + '\n')
    f.write('\n')
f.close()
```
